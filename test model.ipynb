```python
import pickle 
import joblib 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
pd.options.mode.chained_assignment = None
from sklearn.exceptions import FitFailedWarning
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mdl = joblib.load('bank_loan_Mod.pkl')
```

```python
t = pd.read_csv('credit_test.csv')
t
```

```python
t.drop(labels=['Loan ID', 'Customer ID'], axis=1, inplace=True)
t.drop(labels=['Months since last delinquent'], axis=1, inplace=True)
```

```python
t.info()
```

```python
t.drop(t.tail(353).index, inplace=True)
```

```python
t=t.dropna()
```

```python
t['Credit Score'] = t['Credit Score'].apply(lambda val: (val /10) if val>850 else val)
```

```python
t['Years in current job']=t['Years in current job'].str.extract("(\d+)")
t['Years in current job'] = t['Years in current job'].astype(float)
```

```python
t['Employment History'] = t['Years in current job'].apply(lambda
x: "Junior" if x<4 else ("Semi-Senior" if x>4 and x<8 else "Senior"))
```

```python
t.drop(labels=['Years in current job'], axis=1, inplace=True)
```

```python
#t['Home Ownership'].replace({'Home Mortgage':'Own Home'},inplace=True)
t['Home Ownership'].replace({'HaveMortgage':'Home Mortgage'},inplace=True)
```

```python
t['Purpose'].replace({'other':'Other'},inplace=True)
```

```python
t['Tax Liens'] = t['Tax Liens'].apply(lambda x: "No Tax Lien" if
x==0 else "Some Tax Liens")
```

```python
t['Bankruptcies'] =t['Bankruptcies'].apply(lambda x: "No Bankruptcies" if x==0 else "Some Bankruptcies")
```

```python
t['Credit Problems'] = t['Number of Credit Problems'].apply(lambda x: "No Credit Problem" if x==0 else "Some Credit problem")
```

```python
t['Credit Age'] = t['Years of Credit History'].apply(lambda x: "Short Credit Age" if x<5 else ("Good Credit Age" if x>=5 and x<17 else
"Exceptional Credit Age"))
```

```python
t.drop(labels=['Years of Credit History'], axis=1, inplace=True)
```

```python
t.drop(labels=['Number of Credit Problems'], axis=1, inplace=True)
```

```python
for i in t['Current Loan Amount'][t['Current Loan Amount'] == 99999999.0 ].index:
     t.drop(labels=i, inplace=True)
```

```python
t.drop(labels=['Number of Open Accounts'], axis=1, inplace=True)
t.drop(labels=['Current Credit Balance'], axis=1, inplace=True)
t.drop(labels=['Maximum Open Credit'], axis=1, inplace=True)
t.drop(labels=['Monthly Debt'], axis=1, inplace=True)  
```

```python
t
```

```python
# # Encoding categorical data & Feature Scaling

# Select the categorical columns
categorical_subset1 = t[['Term', 'Employment History', 'Home Ownership', 'Purpose','Credit Problems','Bankruptcies','Credit Age','Tax Liens']]
# One hot encode
categorical_subset1 = pd.get_dummies(categorical_subset1)

# Join the dataframe in credit_train
# Make sure to use axis = 1 to perform a column bind
# First I will drop the 'old' categorical datas and after I will join the 'new' one.

t.drop(labels=['Term', 'Employment History', 'Home Ownership', 'Purpose','Credit Problems','Bankruptcies','Credit Age','Tax Liens']
, axis=1, inplace=True)
t = pd.concat([t, categorical_subset1], axis = 1)
```
